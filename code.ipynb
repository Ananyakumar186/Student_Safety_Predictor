```
#importing the libraries and extract the file
import pandas as pd

df=pd.read_csv("/survey.csv")
df.head()
df.info()
```

```
# There is a outlier in dist variable which we can remove by quartile deviation as shown
import numpy as np
uv=3*np.quantile(df.dist_frm_clg,.413)
df.dist_frm_clg[df.dist_frm_clg>uv]=uv
df.head(10)
```

```
#using the dummies library for other indicators variables
df=pd.get_dummies(df)

df=df[:-7]
df.head(10)
```

```
x=df.comfort_rating[:].values

from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x,data=df)
```
